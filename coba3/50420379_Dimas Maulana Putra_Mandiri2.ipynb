### Menghapus Kolom
---

Menghapus kolom dengan fungsi drop dan del

```python
# 4. Menggunakan fungsi drop untuk menghapus kolom ID dan Tahun Lahir

df = df.drop(['ID',"Year_Birth"], axis=1)
df.head()
```

```python
# 5. Menggunakan fungsi del untuk menghapus kolom Dt_Customer

del df['Dt_Customer']
df.head()
```

### Mengolah Nilai Unik dari Suatu Kolom 
---

Proses yang dilakukan terdiri dari menghitung nilai unik dan mengganti nilai unik. 

```python
# 6. Menghitung jumlah nilai unik untuk status pernikahan (marital status)

df['Marital_Status'].value_counts()
```

Nilai unik yang diganti pada kolom pernikahan adalah:
* Married -> Menikah
* Single -> Belum Menikah
* Divorced -> Cerai Hidup
* Widow -> Cerai Mati
* Alone -> Belum Menikah
* YOLO -> Belum Menikah
* Together -> Menikah
* Absurd -> Lainnya

```python
# 7. Mengganti nilai unik pada kolom status pernikahan (marital status)

df['Marital_Status'] = df['Marital_Status'].replace({
    'Married' : 'Menikah'
    ,'Single' : 'Belum Menikah'
    ,'Divorced' : 'Cerai Hidup'
    ,'Widow' : 'Cerai Mati'
    ,'Alone' : 'Belum Menikah'
    ,'YOLO' : 'Belum Menikah'
    ,'Together' : 'Menikah'
    ,'Absurd' : 'Lainnya'
})
df.head()
```

```python
# 8. Menghitung jumlah nilai unik pada kolom status pernikahan yang baru (marital status)

df['Marital_Status'].value_counts()
```

Nilai unik yang diganti pada kolom pendidikan adalah:
* Graduation -> S1
* Master -> S2
* 2n Cycle -> S2
* PhD -> S3
* Basic -> SMA/Sederajat

```python
# 9. Membuat pemetaan nilai lama ke nilai baru untuk kolom Pendidikan (Education)

mapping = {
    'Graduation': 'S1',
    'Master': 'S2',
    '2n Cycle': 'S2',
    'PhD': 'S3',
    'Basic': 'SMA/Sederajat',
}


# 10. Menerapkan pemetaan pada kolom Pendidikan (Education)

df['Education'] = df['Education'].replace(mapping)
df.head()
```

```python
df['Education'].value_counts()
```
