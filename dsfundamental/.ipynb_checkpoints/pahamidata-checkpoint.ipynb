# Data Preprocessing

### Director and Cast (diapus kolomnya) -cara paling gampang

```python
movies.drop(['director','cast'], axis=1, inplace=True) #axis 1 untuk kolom, 0 untuk baris... inplace untuk mengubah data ori-nya
```

```python
movies.isnull().sum()
```

### Country (diisi dengan nilai yang paling sering muncul)

```python
movies['country'] = movies['country'].fillna(movies['country'].mode()[0])
```

```python
movies.isnull().sum()
```

### Rating (diisi dengan data sebenarnya) -upload sendiri cari satu-satu

```python
null_rating = movies['rating'].isnull()
null_rating
```

```python
null_rating_dataframe = movies[null_rating]
null_rating_dataframe
```

```python
rating_replacements = {
    5989: 'PG-13',
    6827: 'TV-14',
    7312: 'R',
    7537: 'TV-MA'
}
```

```python
rating_replacements.items()
```

```python
for id, rate in rating_replacements.items():
    movies.loc[id,'rating'] = rate
```

```python
movies.isnull().sum()
```

### Data Added (mengganti null dengan Not Added)

```python
movies['date_added'] = movies[['date_added']].replace(np.nan, 'Not Added')
#karena replace cuma bisa buat data frame, maka kurung siku 2 agar dari data series berubah menjadi data frame
```

```python
movies.isnull().sum()
```

```python
movies['duration'] = movies[['duration']].replace(np.nan,'37 min')
```

```python
movies.isnull().sum()
```

# Data Visualization

```python
movies['release_year'].value_counts().index[0:15]
```

```python
plt.figure(figsize=(8,6))
sns.set(style='whitegrid')
sns.countplot(y='release_year', data=movies, order=movies['release_year'].value_counts().index[0:15])
plt.title('Release Year', fontsize=12, fontweight='bold')
plt.xlabel('Total')
plt.ylabel('Year')
```

```python
sns.countplot(x='type', data=movies, palette='rocket')
plt.title('TV Show vs Movie', fontsize=12, fontweight='bold')
plt.xlabel('Type')
plt.ylabel('Total')
```

```python
plt.figure(figsize=(12,9))
sns.countplot(x='rating', palette='rocket', data=movies, order=movies['rating'].value_counts().index[0:15])
```

```python
movies['release_month'] = movies['date_added'].apply(lambda x: x.lstrip().split(" ")[0])
```

```python
plt.figure(figsize=(16,8))
sns.countplot(x='release_month', data=movies, order=movies['release_month'].value_counts().index[0:12])
plt.xticks(rotation=45)
```
