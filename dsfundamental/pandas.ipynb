# Series

```python
temperatures = [25,30,32,28,29]
temperatures_series = pd.Series(temperatures)
temperatures_series
```

```python
temperatures_series.describe() #mengembalikan statistik mengenai Series ini
```

```python
temperatures_series.is_unique
```

```python
# kalo mau ngambil std deviasi aja
std_dev = temperatures_series.std()
std_dev
```

# Access Values

```python
for value in temperatures_series:
    print(value)
```

# Modify Values

```python
temperatures_series[1] = 26
temperatures_series
```

```python
temperatures_series = temperatures_series + 1 #bisa melakukan operasi tanpa perulangan
temperatures_series
```

```python
#jika inplace False maka kita tidak mengubah series aslinya
temperatures_series.sort_values(inplace=True)
temperatures_series
```

```python
#drop true itu artinya membuang index lama
temperatures_series.reset_index(drop=True, inplace=True)
temperatures_series
```
